```python
import requests
from bs4 import BeautifulSoup
```

```python
url=requests.get("https://en.wikipedia.org/wiki/List_of_active_Indian_military_aircraft")
print(url.text)
```

```python
soup=BeautifulSoup(url.text,"html.parser")
print(soup.prettify())
```

```python
table=soup.find('table',class_='wikitable')
print(table.text)
```

```python
hl=[]
for i in table.find_all('th'):
    hl.append(i.text.strip())
print(hl)
```

```python
rl=[]
for tr in table.find_all('tr'):
    row=[]
    for j in tr.find_all('td'):
        row.append(j.text.strip())
    if row:
        rl.append(row)
print(rl)

```

```python
import pandas as pd
df=pd.DataFrame(rl)
df.head()
```

```python
header=hl[:6]
header
```

```python
df.columns=header
```

```python
df.to_csv("iaf.csv",index=False)
```
